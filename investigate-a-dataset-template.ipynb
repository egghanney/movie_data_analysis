## Exploratory Data Analysis

>  In this section I will be looking at how movie production have changed over the years with the evolution of technology. I will be using a pivot table function to perform aggregated calculations and proceed to plot my findinds on a bar graph with labels.

### Question 1: How have movie production changed with years?

```python
pd.pivot_table(df, index='release_year', values='original_title', aggfunc='count').plot.bar( ylabel = ('Number of Movies'),title=('Number of movies based of the year'))
def customize(): 
    plt.rcParams["figure.figsize"] = (30, 20)
    plt.rcParams.update({'font.size': 30})
    plt.show()
    return
customize()
```

### Findings 

> From the figure displayed it is evident that number of movies have drastically increased over the years. there has also been an upward consistent trend from 1997, with a drop in just two years being 2010 and 2015. This shows that movie producing in terms of number has been growing since it started. 


> In this part of the analysis, I will be looking at the relationship betweeen budget and revenue over the years. I will be making use of pivot table function and further proceed with a line plot of my findings. 
### Question 2: Determine the relationship between budget and revenue over the years 

### Reason 
I used pivot table because it makes it easier to created aggregated data from a dataset. 

```python
# Creating a pivot table and proceeding with the plot 
pd.pivot_table(df, index='release_year', values=['revenue','budget'], aggfunc='sum').plot.line( ylabel = ('Amount (Trillion)'),title=('Relationship between revenue and budget over the years'))
customize()
```

## Findings 
>  It is clear from the graph that over the years as the revenue for movies kept increasing the budgets also increased, this show that there is a positive coorelation between the two varibles. In other words budget there is an upward trend to the budgets and revenues. 

> Another point worth noting is that, budgets kept increaseing at an increasing rate until after 2010, where it began to show signs of decline, even though not significant at the moment it should be monitored over the next few years.
